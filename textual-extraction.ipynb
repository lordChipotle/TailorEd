# Text Extraction (Bulk) 

```python
import cv2
from pdf2image import convert_from_path
import os
os.listdir(datasetdir)
```

```python
movedir
```

```python
os.listdir(movedir)
```

```python
def extractFullTextBulk(pages,dir):
  prefix_dir = dir
  imgc=1
  text = ""
  for page in pages:
      filename="page_"+str(imgc)+".png"

      page.save(prefix_dir+filename,'PNG')
      
  
      image=cv2.imread(prefix_dir+'page_'+str(imgc)+'.png')
      image_path_in_colab= prefix_dir+'page_'+str(imgc)+'.png'
      extractedInformation = pytesseract.image_to_string(Image.open(image_path_in_colab))

      file=open(prefix_dir+'img'+str(imgc)+'.txt','w') 
      file.write(extractedInformation) 
      file.close()
      file=open(prefix_dir+'img'+str(imgc)+'.txt','r')
      temp=file.read()  
      text = text+temp 
      imgc+=1    
  fullText = text
  file=open(prefix_dir+'fullText.txt','w') 
  file.write(fullText) 
  file.close()
  return fullText
```

```python
from pathlib import Path
import json 

import os

path =datasetdir
#we shall store all the file names in this list
filelist = []

for root, dirs, files in os.walk(path):
	for file in files:
        #append the file name to the list
		filelist.append(os.path.join(root,file))
```

```python
len(filelist)
```

```python
#print all the file names
fullTextDict=allText
for name in filelist:
    if name.endswith('.pdf'):
      
      foldername = Path(name).stem
      newdir = movedir+"/"+foldername+"/"
      if os.path.isdir(newdir):
        pass
      else:
        
        os.mkdir(newdir)
        pages = convert_from_path(name,500)
        s = extractFullTextBulk(pages,newdir)
        fullTextDict[foldername] = s
```

```python
with open('/content/drive/MyDrive/Syllabi [Processed]/allText.json', "w") as outfile:
    json.dump(fullTextDict, outfile)
```
